#### helper function to convert the string representation of lists to list object

```python
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

ast.literal_eval is used to convert string representation of int,list,dic,tuples into corresponding objects

```python
movie_data["genres"]=movie_data["genres"].apply(convert)
```

```python
movie_data.head(1)
```

```python
movie_data["keywords"]=movie_data["keywords"].apply(convert)
```

#### helper function to get top 3 cast members

```python
def get_top3(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
movie_data["cast"]=movie_data["cast"].apply(get_top3)
```

```python
movie_data.iloc[0]["crew"]
```

```python
def get_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i["job"]== "Director":
            l.append(i["name"])
            break
    return l
```

```python
movie_data["crew"].apply(get_director)
```

```python
movie_data["crew"]=movie_data["crew"].apply(get_director)
```

```python
movie_data["overview"]=movie_data["overview"].apply(lambda x:x.split())
```

```python
movie_data["genres"]=movie_data["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_data["keywords"]=movie_data["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movie_data["cast"]=movie_data["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movie_data["crew"]=movie_data["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_data.head(1)
```

```python
movie_data["tags"]=movie_data["genres"]+movie_data["keywords"]+movie_data["overview"]+movie_data["cast"]+movie_data["crew"]
```

```python
movie_data.drop("id",axis=1,inplace=True)
```

```python
final_data=movie_data[["movie_id","title","tags"]]
```

```python
final_data
```

```python
final_data["tags"]=final_data['tags'].apply(lambda x:" ".join(x))
```

```python
final_data["tags"][0]
```

Tokenization: The text is split into individual words or tokens. Punctuation and capitalization are often ignored, and words are converted to lowercase.

```python
final_data["tags"]=final_data["tags"].apply(lambda x:x.lower())
```

```python
final_data["tags"][0]
```

#### Vectorisation:

Each piece of text is represented as a vector, where each element of the vector corresponds to the count of presence of a word in the vocabulary. This results in a high-dimensional, sparse representation.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(final_data["tags"]).toarray()
```

```python
vectors.shape
#so rows represents the no of movies and columns represent the max_feature argument in CountVectorizer
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommended(movie):
    movie_index=final_data[final_data['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(final_data.iloc[i[0]].title)
```

```python
recommended('Avatar')
```
